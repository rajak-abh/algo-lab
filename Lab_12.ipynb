# MATRIX CHAIN MULTIPLICATION Dynamic Programming with table filling.
Problem Description:
Find minimum number of scalar multiplications needed to multiply a chain of matrices..

#Theory
Cost of multiplying two matrices:
If A is p×q and B is q×r, cost = p × q × r

Dynamic Programming idea:
Break problem into smaller subproblems.
Store results in a table to avoid recomputation.

m[i][j] = minimum cost to multiply Ai to Aj

## Algorithm
1. Take array p[] of size n
2. Create table m[n][n]
3. Set m[i][i] = 0
4. Fill table diagonally
5. For each chain length L:
    Compute m[i][j]
    Try all possible k between i and j
6. Choose minimum cost
7. Return m[1][n-1]

```python
import sys

def matrix_chain_order(p):
    n = len(p)
    m = [[0 for _ in range(n)] for _ in range(n)]
    s = [[0 for _ in range(n)] for _ in range(n)]

    # L is chain length
    for L in range(2, n):
        for i in range(1, n - L + 1):
            j = i + L - 1
            m[i][j] = sys.maxsize
            for k in range(i, j):
                cost = m[i][k] + m[k+1][j] + p[i-1]*p[k]*p[j]
                if cost < m[i][j]:
                    m[i][j] = cost
                    s[i][j] = k
    return m, s

def print_optimal_parens(s, i, j):
    if i == j:
        return f"A{i}"
    else:
        return "(" + print_optimal_parens(s, i, s[i][j]) + \
               " x " + print_optimal_parens(s, s[i][j]+1, j) + ")"

def naive_cost(p):
    # Left-to-right multiplication cost
    cost = 0
    matrices = [(p[i], p[i+1]) for i in range(len(p)-1)]
    while len(matrices) > 1:
        A = matrices.pop(0)
        B = matrices.pop(0)
        cost += A[0]*A[1]*B[1]
        matrices.insert(0, (A[0], B[1]))
    return cost

# Example input
p = [5, 10, 3, 12, 5, 50, 6]

m, s = matrix_chain_order(p)
optimal_cost = m[1][len(p)-1]
optimal_parens = print_optimal_parens(s, 1, len(p)-1)
naive = naive_cost(p)

print("Minimum multiplication cost:", optimal_cost)
print("Optimal parenthesization:", optimal_parens)
print("Naive left-to-right cost:", naive)
print("Multiplications saved:", naive - optimal_cost)

print("\nDP Table:")
for row in m[1:]:
    print(row[1:])



```
