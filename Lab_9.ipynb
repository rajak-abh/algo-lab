# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    predecessors = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')
        predecessors[node] = None

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        if min_node is None: # All remaining unvisited nodes are unreachable
            break

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                predecessors[neighbor] = min_node

    return distances, predecessors
```

```python
graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10},
    'C': {'A': 2, 'B': 5, 'D': 3},
    'D': {'B': 10, 'C': 3}
}

distances, predecessors = dijkstra(graph, 'A')

for node, distance in distances.items():
    print(f"Distance from A to {node} = {distance}")
```

### Implementations

Let's implement the requested features: user-defined graph, shortest path display, adding more nodes, and testing with different source nodes.

```python
def get_shortest_path(predecessors, start, target):
    path = []
    current = target
    while current is not None:
        path.insert(0, current)
        current = predecessors[current]
    if path[0] == start:
        return ' -> '.join(path)
    else:
        return f"No path from {start} to {target}"
```

### New Graph with more nodes

Here's an example of a graph with two additional nodes ('E' and 'F'). You can modify this dictionary to define your own graph structure and weights. Please ensure all edge weights are non-negative for Dijkstra's algorithm.


```python
# Define a new graph with additional nodes 'E' and 'F'
new_graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 5, 'D': 10, 'E': 6},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8},
    'D': {'B': 10, 'C': 3, 'E': 7},
    'E': {'B': 6, 'D': 7, 'F': 1},
    'F': {'C': 8, 'E': 1}
}

start_node_1 = 'A'
distances_1, predecessors_1 = dijkstra(new_graph, start_node_1)

print(f"--- Shortest paths from {start_node_1} ---")
for node, distance in distances_1.items():
    if distance == float('inf'):
        print(f"Distance to {node}: Unreachable")
    else:
        path = get_shortest_path(predecessors_1, start_node_1, node)
        print(f"Distance to {node}: {distance}, Path: {path}")
```
